```python
dataset0 = dataset.drop(["Name"],axis=1) # as name is not important in a prediction of goal tally
```

```python
position_arr = []                                #i have seen that many players have a secondary position so to remove that i sliced each strings
for i in dataset0["Position"]:
    if len(i)>2:
        i=i[:2]
    position_arr.append(i)
        
```

```python
position=np.array(position_arr)
#len(position)
# dataset.isnull().sum()
```

```python
dataset1 = dataset0.drop("Position",axis=1)
dataset1["position"]=position
dataset1.head(7)
dataset1["Club"].value_counts().to_dict()
```

```python
dataset1["Club"].unique() #gives unique val
dataset1["Club"].value_counts().to_dict() # gives value count of unq val and making it return as a dictionary
dataset1["Club"].map(dataset1["Club"].value_counts().to_dict()) #maping to replace
```

```python
def NumberEncoder(dataset):    #tested all the needed functions in the prev cell and made a custom func to number encode
    unq_val_arr = dataset.unique()
    dict_val = dataset.value_counts().to_dict()
    new_col = dataset.map(dict_val)
    return np.array(new_col)
    
    
clubs=NumberEncoder(dataset1["Club"])
nation=NumberEncoder(dataset1["Nationality"])
positions = NumberEncoder(dataset1["position"])
```

```python
dataset2 = dataset1.drop(["Club","Nationality"],axis=1)    #added the new columns to the dataset
dataset2["club"] = clubs
dataset2["nation"] = nation
dataset2["position"] = positions
dataset2.head(10)
```

```python
dataset2.corr()["Goals"]

    
```

```python
from sklearn.model_selection import train_test_split

x = dataset2.drop("Goals",axis=1)
y = dataset2["Goals"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)  #Train test splitting
```

```python
from sklearn.linear_model import LinearRegression
linregg = LinearRegression()
linregg.fit(x_train,y_train) # training the model in linear regg
```

```python
from sklearn.metrics import r2_score 
y_pred = linregg.predict(x_test)  #fitting test data to model
acc = r2_score(y_pred, y_test)*100
print(f"the accuracy of the model is {acc}%")
```

Using random forest algorithm

```python
from sklearn.ensemble import RandomForestRegressor
rand_f = RandomForestRegressor()
```

```python
rand_f.fit(x_train,y_train) # training the model in rf regg
```

```python
y_pred = rand_f.predict(x_test)  #fitting test data to model
acc2 =r2_score(y_pred,y_test)*100
print(f"the accuracy of the model is {acc2}%")
```
